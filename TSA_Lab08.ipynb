```python
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
```

```python
def split_sequence(sequence,steps):
  X,y = list(), list()
  for start in range(len(sequence)):
    index_end = start+steps
    if index_end>len(sequence)-1:
      break
    sequence_x, sequence_y = sequence[start: index_end],sequence[index_end]
    X.append(sequence_x)
    y.append(sequence_y)
  return array(X), array(y)
```

```python
import pandas as pd
df = pd.read_csv('/content/Daily_Electricity_Consumption.csv')
```

```python
df.head()
```

```python
import numpy as np
usage = np.array(df['usage'])
```

```python
steps = 3
X,y = split_sequence(usage,steps)

print("The input raw sequence on which we will test put CNN model is : ")
print(usage)

print("\n\nThe features (X), lagged values of time series, for our CNN model : ")
print(X)

print("\n\nThe respective observed values (y) of the sequence for training in CNN model:")
print(y)
```

```python
features = 1
X = X.reshape((X.shape[0],X.shape[1],features))

print("After Shaping the shape of input X is : ")
print(X.shape)

print("The final form of input feature matrix X : ")
print(X)
```

## Model Building

```python
model = Sequential()
model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(steps,features)))
model.add(MaxPooling1D(pool_size=2))
model.add(Flatten())
model.add(Dense(100, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss=['mse'],metrics=['mae','mape'])
```

```python
model.fit(X,y,epochs=1000, verbose=True)
```

```python
model.summary()
```

```python
x_input = array([6,5,8])
x_input = x_input.reshape((1,steps,features))
y_pred = model.predict(x_input,verbose=0)
print("The next predicted value for the sequence is ")
print(y_pred)
```

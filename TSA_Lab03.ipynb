# Selected DataSet

```python
df = pd.read_csv('/content/D202.csv')
df.head()
```

Type, Units, End Time, Notes are redundant columns so we drop them

```python
df.drop(columns=["TYPE", "UNITS", "END TIME", "NOTES"], errors="ignore",inplace=True)
```

```python
df["COST"] = df["COST"].str.replace("$", "", regex=False).astype(float)
```

We attach the Timestamp along with Date

```python
add_prefix_lambda = lambda start_time: "0"+ start_time if int(start_time.split(":")[0]) < 12 else start_time
df["START TIME"] = df["START TIME"].apply(add_prefix_lambda)
```

```python
df["Date"] = df["DATE"] + " " + df["START TIME"]
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df[["Date", "USAGE", "COST"]]
df.columns = ["Date", "usage", "cost"]
```

```python
df.head()
```

We take the aggregate and sum the usage and cost for monthly dataset

```python
monthly_usage_df = df.resample("1M", on="Date").sum()
```

```python
monthly_usage_df.head()
```

```python
monthly_usage_df.shape
```

```python
monthly_usage_df.agg({'usage': ['count','min', 'max', 'mean']})
```

We take the aggregate and sum the usage and cost for Daily dataset

```python
daily_usage_df = df.resample("1D", on="Date").sum()
```

```python
daily_usage_df.shape
```

```python
daily_usage_df.head(10)
```

```python
daily_usage_df.agg({'usage': ['count','min', 'max', 'mean']})
```

```python
daily_usage_df[(daily_usage_df['usage'] >=5.00) & (daily_usage_df['usage'] <=15.00)]
```

```python
daily_usage_df.to_csv('Daily_Electricity_Consumption.csv')
```
